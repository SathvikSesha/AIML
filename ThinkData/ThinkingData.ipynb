```python
import json
```

```python
# Load data
def loadData(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data    
```

```python
data = loadData("store_data.json")
```

```python
def printdata(data):
    for user in data:
        print(user)
```

```python
def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"Five":5}
    for user in data:
        x = str(user['rating']).strip().lower()
        if(x in text_to_num):
            user['rating'] = text_to_num[x]
```

```python
clean_data(data)
```

```python
printdata(data)
```

```python
def missing_replace(data):
    for user in data:
        x = user.get('age')
        if(x == None):
            user['age'] = None
```

```python
missing_replace(data)
printdata(data)
```

```python
def remove_duplicates(data):
    s = set()
    l = []
    for user in data:
        name_key = user['name'].strip().lower()
        if name_key in s:
            continue
        s.add(name_key)
        l.append(user)
    return l
```

```python
data = remove_duplicates(data)
```

```python
## Getting meaningful insights from data
# Avg rating 
def avg_rating(data):
    total = 0
    for user in data:
        total+=float(user['rating'])
    print(f'Avg rating : {total/len(data)}')
    return total/len(data)
```

```python
x = avg_rating(data)
x
```

```python
def user_rated_poor(data):
    total_user = 0
    for user in data:
        if float(user['rating']) < 3:
            total_user+=1
    return (total_user/len(data))*100
```

```python
poorRating = user_rated_poor(data)
poorRating
```

```python
# build recommandations
def get_recommandation(data):
    recommandations = []
    for user in data:
        if float(user['rating']) >=4:
            brand = "Apple"
        else:
            brand = "Samsung"
        recommandations.append(
            {
                "name" : user['name'].strip(),
                "recommand" : brand
            }
        )
    return recommandations
```

```python
rec = get_recommandation(data)
for user in rec:
    print(f"{user['name']} gets recommanded {user['recommand']}")
```
